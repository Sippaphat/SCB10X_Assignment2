```python
url = "https://krupaga.wordpress.com/category/แบบทดสอบรายวิชาการประก/"

downloaded = fetch_url(url)
result = extract(downloaded)

```

```python
import re

def extract_exam(text):
    # Regular expressions for question, choices, and answer patterns
    question_pattern = r"(\d+\.)(.+)"
    choice_pattern = r"([ก-ฮ]\.)(.+)"
    answer_pattern = r"เฉลย(.+)"

    questions = []
    current_question = {}

    for line in text.splitlines():
        line = line.strip()  # Remove leading/trailing whitespace

        # Match questions
        question_match = re.match(question_pattern, line)
        if question_match:
            if current_question:  # Save previous question if any
                questions.append(current_question)
            current_question = {
                "question": question_match.group(2),
            }

        # Match choices
        choice_match = re.match(choice_pattern, line)
        if choice_match:
            current_question[choice_match.group(1)] = choice_match.group(2)

        # Match answer
        answer_match = re.match(answer_pattern, line)
        if answer_match:
            current_question["answer"] = answer_match.group(1).strip()

    # Add the last question
    if current_question:
        questions.append(current_question)

    
    return questions
    
# Replace this with your actual scraped text
text = result
exam_data = extract_exam(text)
for q in exam_data:
    q["subject"] = "อาหารไทย"
    print(q)
```

```python
with jsonlines.open("exam_data.jsonl", mode='w') as writer:
    for item in exam_data:  # exam_data คือ list ของ dictionaries ที่เก็บข้อมูลข้อสอบ
        writer.write(item)
```

```python
import re

def extract_exam_questions(scraped_text):
    """Extracts exam questions and answers from scraped text.

    Args:
        scraped_text: The text obtained from web scraping.

    Returns:
        A list of dictionaries, each containing a question, choices, and answer.
    """

    question_pattern = re.compile(
        r"(\d+\.)(.+?)(?=ก\. |ข\. |ค\. |ง\. |\d+\. |$)",
        re.DOTALL  # Allow '.' to match newlines
    )
    choice_pattern = re.compile(r"(ก\. |ข\. |ค\. |ง\. )(.+?)(?=\n|$)")
    answer_pattern = re.compile(r"เฉลย\s*(.+?)\n")

    questions = []
    for match in question_pattern.finditer(scraped_text):
        question_number, question_text = match.groups()

        choices = []
        for choice_match in choice_pattern.finditer(question_text):
            choice_letter, choice_text = choice_match.groups()
            choices.append((choice_letter, choice_text.strip()))

        answer_match = answer_pattern.search(question_text)
        answer = answer_match.group(1).strip() if answer_match else None

        questions.append({
            "question_number": question_number,
            "question_text": question_text.strip(),
            "choices": choices,
            "answer": answer
        })

    return questions
```

```python
# (Assuming you have the 'scraped_text' variable containing your scraped exam content)
scraped_text = result
questions = extract_exam_questions(scraped_text)

for question in questions:
    print("Question:", question["question_number"], question["question_text"])
    for choice_letter, choice_text in question["choices"]:
        print(f"  {choice_letter} {choice_text}")
    print("Answer:", question["answer"])
    print("---")
```
